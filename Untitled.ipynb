# Maldives population

```python
import pandas as pd

df = pd.read_csv('mdv_admpop_adm1_2022.csv')  # file name should match the uploaded one
df.head()
```

```python
df_subset = df[['ADM2_EN', 'F_TL', 'M_TL']]
df_subset = df_subset.dropna()  # remove any missing data if present
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Setup
cities = df_subset['ADM2_EN']
female_pop = df_subset['F_TL']
male_pop = df_subset['M_TL']
x = np.arange(len(cities))  # x locations for bars

# Plotting
plt.figure(figsize=(15, 6))
bar_width = 0.4

plt.bar(x - bar_width/2, female_pop, width=bar_width, label='Female', color='salmon')
plt.bar(x + bar_width/2, male_pop, width=bar_width, label='Male', color='skyblue')

# Labels & Title
plt.xticks(x, cities, rotation=90)
plt.xlabel("City / Atoll")
plt.ylabel("Population")
plt.title("Male vs Female Population by City (ADM2 Level)")
plt.legend()
plt.tight_layout()
plt.grid(axis='y')

plt.show()
```

```python
df['M_per_100_F'] = df['M_TL'] / df['F_TL'] * 100

plt.figure(figsize=(14, 5))
plt.bar(df['ADM2_EN'], df['M_per_100_F'], color='mediumpurple')
plt.axhline(100, color='gray', linestyle='--')  # reference line
plt.xticks(rotation=90)
plt.ylabel('Males per 100 Females')
plt.title('Gender Ratio (Male per 100 Female) by City')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and extract age group columns
df = pd.read_csv('mdv_admpop_adm1_2022.csv')
male_cols = [col for col in df.columns if col.startswith('M_') and col != 'M_TL']
female_cols = [col for col in df.columns if col.startswith('F_') and col != 'F_TL']

# Sum population for each age group
male_sums = df[male_cols].sum()
female_sums = df[female_cols].sum()

# Age group labels (remove 'M_' / 'F_')
age_labels = [col[2:] for col in male_cols]

# Reverse for pyramid style
male_sums = -male_sums.values  # negative for left side
female_sums = female_sums.values[::-1]
male_sums = male_sums[::-1]

# Plot
plt.figure(figsize=(10, 8))
y = range(len(age_labels))

plt.barh(y, male_sums, color='#4D6CFA', label='Male')
plt.barh(y, female_sums, color='#FF69B4', label='Female')

plt.yticks(y, age_labels[::-1])
plt.xlabel("Population")
plt.title("Population Pyramid by Age Group and Gender")
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
